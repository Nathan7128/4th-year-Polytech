_Veuillez représenter le diagramme de classes du simple programme suivant (un morpion)_

```python
class Player:
    def __init__(self, playerName):
        self.username = playerName

class Board:
    def __init__(self):
        self.board = [['.' for x in range(0, 3)] for y in range(0, 3)]

    def printBoard(self):
        for y in range(0, len(self.board)):
            for x in range(0, len(self.board[y])):
                print(self.board[y][x], end = "|")
            print("")
    
    def setValue(self, y, x, param):
        self.board[y][x] = param

    def solve(self):
        
        # horizontal
        for y in range(0, len(self.board)):
            state = self.board[y][0]
            flag = True
            if(state == '.'):
                continue
            for x in range(1, len(self.board[y])):
                if(state != self.board[y][x]):
                    flag = False
            if(flag):
                return True
        
        # vertical
        for x in range(0, len(self.board[0])):
            state = self.board[0][x]
            flag = True
            if(state == '.'):
                continue
            for y in range(1, len(self.board[0])):
                if(state != self.board[y][x]):
                    flag = False
            if(flag):
                return True
        
        # diagonal
        if(self.board[1][1] != '.' and self.board[0][0] == self.board[1][1] and self.board[1][1] == self.board[2][2]):
            return True
        if(self.board[1][1] != '.' and self.board[0][2] == self.board[1][1] and self.board[1][1] == self.board[2][0]):
            return True
        
        # pas de vainqueur
        return False

    # Check s'il reste des cases non remplies dans la grille
    def isValid(self):
        for y in range(0, len(self.board)):
            for x in range(0, len(self.board[0])):
                if(self.board[y][x] == '.'):
                    return True
        return False
    
    # Check si le move d'un jouer est légal.
    def validMove(self, y, x):
        if(self.board[y][x] != '.'):
            return False
        return True

class Game:
    def __init__(self, user1, user2):
        self.player1 = user1
        self.player2 = user2
        self.choice = True # True shows player1 has chance to go first
        self.board = Board()
    
    def gameInstance(self):
        self.board.printBoard()
    
    def play(self):
        if(not self.board.isValid()):
            print("DRAW")
            return
        if(self.choice):
            print("Player 1: "+ self.player1.username + " Move")
            play1Y, play1X = list(map(int, input().strip().split()))
            if(self.board.validMove(play1Y, play1X)):
                self.board.setValue(play1Y, play1X, 'X')
                self.gameInstance()
                if(self.board.solve()):
                    print("Player 1:" + self.player1.username + " Won")
                    return
                self.choice = False
                self.play()
            else:
                print("Invalid Move")
                self.play()
        else:
            print("Player 2: "+ self.player2.username + " Move")
            play2Y, play2X = list(map(int, input().strip().split()))
            if(self.board.validMove(play2Y, play2X)):
                self.board.setValue(play2Y, play2X, 'O')
                self.gameInstance()
                if(self.board.solve()):
                    print("Player 2:" + self.player2.username + " Won")
                    return
                self.choice = True
                self.play()
            else:
                print("Invalid Move")
                self.play()



player1 = Player("Kévin Merle")
player2 = Player("Olivier Merle")
newGame = Game(player1, player2)
newGame.gameInstance()
newGame.play()
```
