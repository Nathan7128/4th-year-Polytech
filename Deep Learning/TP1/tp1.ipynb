```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
seed = 7
np.random.seed(seed)

N = 1000
X_train = np.linspace(0.0, 1.0, N)
y_train = np.sin(2*np.pi*X_train)

X_val = np.random.rand(100)
X_val = np.sort(X_val)
y_val = np.sin(2*np.pi*X_val)
```

```python
plt.plot(X_val, y_val, "r-")

plt.plot(X_train, y_train, 'b-')

plt.title("Training and validation data")

plt.Text(0.5, 1.0, "Training and validation data")
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
N1 = 50
N2 = 16

model=Sequential()
model.add(Dense(N1,input_dim=1, kernel_initializer='uniform', activation='relu'))
model.add(Dense(N2, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform'))

# model.summary()

model.compile(optimizer='rmsprop',loss='mse',metrics=['mse'])

model.fit(X_train,y_train,epochs=400, batch_size=32, validation_data=(X_val,y_val))

y_p = model.predict(X_val)
plt.plot(X_val, y_p, "r-")
plt.plot(X_val, y_val, "b-")
plt.title("Prediction and data")

plt.text(0.5, 1.0, "Prediction and data")
```
